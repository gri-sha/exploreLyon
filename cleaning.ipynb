```python
import pandas as pd
from pprint import pprint
```

```python
df = pd.read_csv("data/flickr_data2.csv")
df.head(30)
```

```python
df.info()
```

```python
# remove the unnecessary space from column names
df.columns = df.columns.str.strip()  # VERY IMPORTANT
df.info()
```

```python
# We can't conclude precisely what these unnamed columns represent: taken or upload year, so we will just drop them
mask = (~df["Unnamed: 16"].isna()) | (~df["Unnamed: 17"].isna()) | (~df["Unnamed: 18"].isna())
df[mask].head(10)
```

```python
# Drop unnamed columns (they contain mostly NaN values)
df.drop(columns=["Unnamed: 16", "Unnamed: 17", "Unnamed: 18"], inplace=True)
df.info()
```

```python
print(f"Rows removed: {df.duplicated().sum()}")
df = df.drop_duplicates()
print(f"New dataframe length: {len(df)}\n")
df.info()
```

```python
# choose only rows with non-NaN date components
# (the following conversion to datetime will fail otherwise)
mask = (
    df["date_upload_minute"].isna()
    | df["date_upload_hour"].isna()
    | df["date_upload_day"].isna()
    | df["date_upload_month"].isna()
    | df["date_upload_year"].isna()
    | df["date_taken_minute"].isna()
    | df["date_taken_hour"].isna()
    | df["date_taken_day"].isna()
    | df["date_taken_month"].isna()
    | df["date_taken_year"].isna()
)
df = df[~mask]
print(f"Rows removed: {mask.sum()}")
print(f"New dataframe length: {len(df)}\n")
df.isna().sum()
```

```python
# change to datetime dtype
# consult this https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html
# the dates that are invalid will be converted to NaT
df["date_upload"] = pd.to_datetime(
    df[["date_upload_year", "date_upload_month", "date_upload_day", "date_upload_hour", "date_upload_minute"]].rename(
        columns={
            "date_upload_year": "year",
            "date_upload_month": "month",
            "date_upload_day": "day",
            "date_upload_hour": "hour",
            "date_upload_minute": "minute",
        }
    ),
    errors="coerce",  # Invalid dates become NaT
)
df["date_taken"] = pd.to_datetime(
    df[["date_taken_year", "date_taken_month", "date_taken_day", "date_taken_hour", "date_taken_minute"]].rename(
        columns={
            "date_taken_year": "year",
            "date_taken_month": "month",
            "date_taken_day": "day",
            "date_taken_hour": "hour",
            "date_taken_minute": "minute",
        }
    ),
    errors="coerce",  # Invalid dates become NaT
)

# Drop rows with NaT values
print(f"Rows removed: {df[df['date_upload'].isna() | df['date_taken'].isna()].shape[0]}")
df = df.dropna(subset=["date_upload", "date_taken"])
print(f"New dataframe length: {len(df)}\n")

# drop the now unnecessary columns
df.drop(
    columns=[
        "date_upload_year",
        "date_upload_month",
        "date_upload_day",
        "date_upload_hour",
        "date_upload_minute",
        "date_taken_year",
        "date_taken_month",
        "date_taken_day",
        "date_taken_hour",
        "date_taken_minute",
    ],
    inplace=True,
)
df.info()
print()
df.isna().sum()
```

```python
# Check that date_taken is before date_upload
mask = df['date_taken'] <= df['date_upload']
df = df[mask]
print(f"Rows removed: {(~mask).sum()}")
print(f"New dataframe length: {len(df)}\n")

df.info()
```

```python
# Sanity check for realistic dates
mask = (df['date_upload'].dt.year >= 2004) & (df['date_taken'].dt.year >= 1826)  # first photo ever taken in 1826, Flickr founded in 2004
df = df[mask]
print(f"Rows removed: {(~mask).sum()}")
print(f"New dataframe length: {len(df)}\n")

df.info()
```

```python
# choose lat/long bounds for Lyon
mask = df["lat"].between(44.65, 45.9) & df["long"].between(4.70, 5.12)
df = df[mask]
print(f"Rows removed: {(~mask).sum()}")
print(f"New dataframe length: {len(df)}\n")

df.info()
```

```python
df["tags"] = df["tags"].fillna("")
# Split by comma, convert to lowercase, and strip whitespace and remove empty tags
df["tags"] = df["tags"].str.split(",").apply(lambda x: [tag.strip().lower() for tag in x if tag.strip()])

df["tags"].head(10)
```

```python
df["title"] = df["title"].fillna("")
```

```python
df.info()
```

```python
df["url"] = df.apply(lambda row: f"https://www.flickr.com/photos/{row['user']}/{row['id']}", axis=1)
pprint(df["url"][:10].to_list())
```
